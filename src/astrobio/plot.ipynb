```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
file = r'flare.csv'

df = pd.read_csv(file)
print(df)
```

```python
fig, ax = plt.subplots()
ax.scatter(df['C02+'], df['Altitude'], label = 'C02+', s=2)
ax.scatter(df['O2+'], df['Altitude'], label = 'O2+', s=2)
plt.xscale('log')
ax.legend()
ax.set_ylabel('Altitude')
ax.set_xlabel('Concentration')

plt.plot()
```

```python
oxy = []
carb = []
last = df.iloc[0]
for idx, row in df.iterrows():
    oxy_inc = (row['O2+'] - last['O2+']) / np.abs(last['O2+']) * 100
    carb_inc = (row['C02+'] - last['C02+']) / np.abs(last['C02+']) * 100
    last = row
    
    oxy.append(oxy_inc)
    carb.append(carb_inc)
```

```python
fig, ax = plt.subplots()
print(len(oxy), len(df))
ax.plot(carb, df['Altitude'], label = 'C02+')
ax.plot(oxy, df['Altitude'], label = 'O2+')
# plt.xscale('log')
ax.legend()
ax.set_ylabel('Altitude')
ax.set_xlabel('Concentration')
ax.set_ylim((150, 250))
plt.plot()
```
